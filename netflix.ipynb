```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Start coding!
netflix_df = pd.read_csv("netflix_data.csv")    #Loading CSV File
#print(netflix_df[0:4])                         #Print 1st 4 rows to check the import is correct or not

netflix_subset = netflix_df[netflix_df["type"] == "Movie"]   #Taking out only movies from whole data
# print(netflix_subset)                         #To check subset

netflix_movies = netflix_subset[['title', 'country', 'genre', 'release_year', 'duration']] #keeping these category to investigate
#print(netflix_movies)  

short_movies = netflix_movies[netflix_movies.duration < 60]   #finding movies with duration < 60mins 
#print(short_movies)

colors = []                                         #empty list for colors
for label, row in netflix_movies.iterrows():        # Iterate over rows of netflix_movies
    if row ['genre'] == "Children":                    #assign different colors
        colors.append("red")
    elif row ['genre'] == "Documentaries":
        colors.append("green")
    elif row ['genre'] == "Stand-Up":
        colors.append("blue")
    else:
        colors.append("white")

# colors[:4]

fig = plt.figure(figsize=(12,8))         #New figure and figure style 

plt.scatter(netflix_movies.release_year, netflix_movies.duration, c=colors) #scatter plot w release year and duration

plt.title("Movie Duration by Year of Release")      #graph title
plt.xlabel("Release year")      #label for x-axis
plt.ylabel("Duration (min)")    #label for y-axis

plt.show()  #print graph

answer = "maybe"
```
